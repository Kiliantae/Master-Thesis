```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))
titles = ["a)","b)","c)","d)","e)","f)"]
i=0
j=0
for p in [10,20,100]:
    df = pd.read_excel('data/Metainformation/Raw_data_zoom{}_224_224(12)_clust.xlsx'.format(p))
    #df = pd.read_csv("data/processed_data/{}zoom_224_224(12)_clust/".format(p)+'{}zoom_LBP_hist.csv'.format(p))
    y =df[["Bruchart-id"]]
    y = np.array(y)
    y_arr = y.tolist()
    z =df[["Material_cluster"]]
    z = np.array(z)
    z_arr = z.tolist()
    X = df.drop(["Bruchart-id","Material_cluster"], axis=1).to_numpy()

    y_0 = y_arr.count([0])
    y_1 = y_arr.count([1])
    y_count = np.array([y_0,y_1])
    y_labels = ["fatique","brittle"]
    #y = np.array([35, 25, 25, 15])
    #mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

    axes[0,i].pie(y_count, labels = y_labels, autopct='%1.1f%%')
    axes[0, i].set_title(titles[j])
    j=j+1
    
    z_0 = z_arr.count([0])
    z_1 = z_arr.count([1])
    z_2 = z_arr.count([2])
    z_3 = z_arr.count([3])
    z_count = np.array([z_0,z_1,z_2,z_3])
    z_labels = ["0","1","2","3"]

    axes[1,i].pie(z_count, labels = z_labels, autopct='%1.1f%%')
    axes[1, i].set_title(titles[j])
    i=i+1
    j=j+1
plt.show() 
```
